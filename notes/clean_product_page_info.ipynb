```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set_theme(style="whitegrid", context="notebook")
plt.rcParams["figure.figsize"] = (10, 5)

DATA_PATH = Path("../input/product_page_info_bath_bombs_all.csv")
df_raw = pd.read_csv(DATA_PATH, low_memory=False)
```

```python
df_raw.columns
```

```python
columns_to_keep = [
    "asin",
    "asin_in_html",
    "title",
    "brand",
    "manufacturer",
    "price",
    "per_unit_price_text",
    "unit_text",
    "per_unit_price",
    "unit_num",
    "label_unit",
    "label_unit_num",
    "number_of_items",
    "item_weight",
    "size",
    "badge_label",
    "image_num",
    "variation_quantity",
    "product_dimensions",
    "feature",
    "product_description",
]

df = df_raw[columns_to_keep].copy()
df.head()
```

```python
asin_norm = df["asin"].astype("string").str.strip()
asin_html_norm = df["asin_in_html"].astype("string").str.strip()

both_present = asin_norm.notna() & asin_html_norm.notna()
equal_mask = both_present & (asin_norm == asin_html_norm)
mismatch_mask = both_present & (asin_norm != asin_html_norm)

summary = pd.Series(
    {
        "total_rows": len(df),
        "asin_missing": asin_norm.isna().sum(),
        "asin_in_html_missing": asin_html_norm.isna().sum(),
        "both_present_and_equal": equal_mask.sum(),
        "both_present_but_different": mismatch_mask.sum(),
    }
)
print(summary)

df.loc[mismatch_mask, ["asin", "asin_in_html", "title"]]
```

```python
df_with_n_items = df[df["number_of_items"].notna()].copy()
print(f"Rows with number_of_items not null: {len(df_with_n_items):,} / {len(df):,}")
df_with_n_items.head()
```
